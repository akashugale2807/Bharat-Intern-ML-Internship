```python
winedf = winedf.drop('total sulfur dioxide', axis=1)
```

```python
winedf.head(5)
```

```python
winedf['best quality'] = [1 if x > 5 else 0 for x in winedf.quality]
winedf.replace({'white': 1, 'red': 0}, inplace=True)
winedf.head(5)
```

```python
winedf['type'].value_counts()
```

```python
winedf['best quality'].value_counts()
```

```python
winedf = winedf.drop('quality', axis=1)
winedf.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics

features = winedf.drop(['best quality'], axis=1)
target = winedf['best quality']
 
xtrain, xtest, ytrain, ytest = train_test_split(
    features, target, test_size=0.2, random_state=40)
 
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
norm = MinMaxScaler()
xtrain = norm.fit_transform(xtrain)
xtest = norm.transform(xtest)
```

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import plot_confusion_matrix, accuracy_score, confusion_matrix, classification_report

logreg = LogisticRegression()
logreg.fit(xtrain, ytrain)

train_predictions = logreg.predict(xtrain)
test_predictions = logreg.predict(xtest)

train_accuracy = accuracy_score(ytrain, train_predictions)
test_accuracy = accuracy_score(ytest, test_predictions)

plt.figure(figsize=(1,1))
plot_confusion_matrix(logreg, xtrain, ytrain)
plt.title('Training Confusion Matrix')
plt.show()

plt.figure(figsize=(1,1))
plot_confusion_matrix(logreg, xtest, ytest)
plt.title('Testing Confusion Matrix')
plt.show()

classification_rep = classification_report(ytest, test_predictions)

print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)

print("\nClassification Report:")
print(classification_rep)
```

```python
# Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier()
rf_classifier.fit(xtrain, ytrain)

train_predictions = rf_classifier.predict(xtrain)
test_predictions = rf_classifier.predict(xtest)

train_accuracy = accuracy_score(ytrain, train_predictions)
test_accuracy = accuracy_score(ytest, test_predictions)

plt.figure(figsize=(1,1))
plot_confusion_matrix(rf_classifier, xtrain, ytrain)
plt.title('Training Confusion Matrix')
plt.show()

plt.figure(figsize=(1,1))
plot_confusion_matrix(rf_classifier, xtest, ytest)
plt.title('Testing Confusion Matrix')
plt.show()

classification_rep = classification_report(ytest, test_predictions)

print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)

print("\nClassification Report:")
print(classification_rep)
```

```python
# Support Vector Machines (SVM)

from sklearn.svm import SVC

svm_classifier = SVC()
svm_classifier.fit(xtrain, ytrain)

train_predictions = svm_classifier.predict(xtrain)
test_predictions = svm_classifier.predict(xtest)

train_accuracy = accuracy_score(ytrain, train_predictions)
test_accuracy = accuracy_score(ytest, test_predictions)

plt.figure(figsize=(1,1))
plot_confusion_matrix(svm_classifier, xtrain, ytrain)
plt.title('Training Confusion Matrix')
plt.show()

plt.figure(figsize=(1,1))
plot_confusion_matrix(svm_classifier, xtest, ytest)
plt.title('Testing Confusion Matrix')
plt.show()

classification_rep = classification_report(ytest, test_predictions)

print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)

print("\nClassification Report:")
print(classification_rep)
```

```python
# K-Nearest Neighbors (KNN)

from sklearn.neighbors import KNeighborsClassifier

knn_classifier = KNeighborsClassifier()
knn_classifier.fit(xtrain, ytrain)

train_predictions = knn_classifier.predict(xtrain)
test_predictions = knn_classifier.predict(xtest)

train_accuracy = accuracy_score(ytrain, train_predictions)
test_accuracy = accuracy_score(ytest, test_predictions)

plt.figure(figsize=(1,1))
plot_confusion_matrix(knn_classifier, xtrain, ytrain)
plt.title('Training Confusion Matrix')
plt.show()

plt.figure(figsize=(1,1))
plot_confusion_matrix(knn_classifier, xtest, ytest)
plt.title('Testing Confusion Matrix')
plt.show()

classification_rep = classification_report(ytest, test_predictions)

print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)

print("\nClassification Report:")
print(classification_rep)
```

```python
# Gradient Boosting Classifier

from sklearn.ensemble import GradientBoostingClassifier

gb_classifier = GradientBoostingClassifier()
gb_classifier.fit(xtrain, ytrain)

train_predictions = gb_classifier.predict(xtrain)
test_predictions = gb_classifier.predict(xtest)

train_accuracy = accuracy_score(ytrain, train_predictions)
test_accuracy = accuracy_score(ytest, test_predictions)

plt.figure(figsize=(1,1))
plot_confusion_matrix(gb_classifier, xtrain, ytrain)
plt.title('Training Confusion Matrix')
plt.show()

plt.figure(figsize=(1,1))
plot_confusion_matrix(gb_classifier, xtest, ytest)
plt.title('Testing Confusion Matrix')
plt.show()

classification_rep = classification_report(ytest, test_predictions)

print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)

print("\nClassification Report:")
print(classification_rep)
```
